Vader Sentiment Analysis

```python
def vader_comparison(texts):
    #headers = ['pos','neg','neu','compound']
    #print("Name\t",'  pos\t','neg\t','neu\t','compound')
    
    analyzer = SentimentIntensityAnalyzer()
    sentences = sent_tokenize(texts)
    
    pos=compound=neu=neg=0
    for sentence in sentences:
        vs = analyzer.polarity_scores(sentence)
        pos+=vs['pos']/(len(sentences))
        compound+=vs['compound']/(len(sentences))
        neu+=vs['neu']/(len(sentences))
        neg+=vs['neg']/(len(sentences))
        # print('%-10s'%name,'%1.2f\t'%pos,'%1.2f\t'%neg,'%1.2f\t'%neu,'%1.2f\t'%compound)
    return [pos, neg, compound, neu]
```

```python
vader_comparison(text1)
```

```python
df['vader'] = df['reviews'].apply(vader_comparison)
```

```python
df.head()
```

```python
def get_pos(vader_list):
    return vader_list[0]

def get_neg(vader_list):
    return vader_list[1]

def get_com(vader_list):
    return vader_list[2]

def get_neu(vader_list):
    return vader_list[3]

df['positive'] = df['vader'].apply(get_pos)
df['negative'] = df['vader'].apply(get_neg)
df['compound'] = df['vader'].apply(get_com)
df['neutral'] = df['vader'].apply(get_neu)
```

```python
df['pos-neg'] = df['positive'] - df['negative']
df.head()
```

```python
df = df.rename(columns = {'positive':'vader_positive'})
df = df.rename(columns = {'negative':'vader_negative'})
df = df.rename(columns = {'pos-neg':'vader_pos_neg'})
```

Simple Sentiment Analysis Using NRC data

```python
def get_nrc_data():
    nrc = "/Users/youngcj/Desktop/IEOR 4523 Data Analytics/Codes/Week 7 Text Mining/data/NRC-emotion-lexicon-wordlevel-alphabetized-v0.92.txt"
    count=0
    emotion_dict=dict()
    with open(nrc,'r') as f:
        all_lines = list()
        for line in f:
            if count < 46:
                count+=1
                continue
            line = line.strip().split('\t')
            if int(line[2]) == 1:
                if emotion_dict.get(line[0]):
                    emotion_dict[line[0]].append(line[1])
                else:
                    emotion_dict[line[0]] = [line[1]]
    return emotion_dict
```

```python
emotion_dict = get_nrc_data()
emotion_dict
```

```python
def NRC_emotion_analyzer(text, emotion_dict = emotion_dict):
    #Set up the result dictionary
    emotions = {x for y in emotion_dict.values() for x in y}
    emotion_count = dict()
    for emotion in emotions:
        emotion_count[emotion] = 0

    #Analyze the text and normalize by total number of words
    total_words = len(text.split())
    for word in text.split():
        if emotion_dict.get(word):
            for emotion in emotion_dict.get(word):
                emotion_count[emotion] += 1/len(text.split())
    return emotion_count
```

```python
df['NRC'] = df['reviews'].apply(NRC_emotion_analyzer)
```

```python
def get_NRC_pos(NRC_dict):
    return NRC_dict['positive']

def get_NRC_sur(NRC_dict):
    return NRC_dict['surprise']

def get_NRC_ant(NRC_dict):
    return NRC_dict['anticipation']

def get_NRC_sad(NRC_dict):
    return NRC_dict['sadness']

def get_NRC_neg(NRC_dict):
    return NRC_dict['negative']

def get_NRC_tru(NRC_dict):
    return NRC_dict['trust']

def get_NRC_ang(NRC_dict):
    return NRC_dict['anger']

def get_NRC_joy(NRC_dict):
    return NRC_dict['joy']

def get_NRC_dis(NRC_dict):
    return NRC_dict['disgust']

def get_NRC_fea(NRC_dict):
    return NRC_dict['fear']

df['NRC_positive'] = df['NRC'].apply(get_NRC_pos)
df['NRC_surprise'] = df['NRC'].apply(get_NRC_sur)
df['NRC_anticipation'] = df['NRC'].apply(get_NRC_ant)
df['NRC_sadness'] = df['NRC'].apply(get_NRC_sad)
df['NRC_negative'] = df['NRC'].apply(get_NRC_neg)
df['NRC_trust'] = df['NRC'].apply(get_NRC_tru)
df['NRC_anger'] = df['NRC'].apply(get_NRC_ang)
df['NRC_joy'] = df['NRC'].apply(get_NRC_joy)
df['NRC_disgust'] = df['NRC'].apply(get_NRC_dis)
df['NRC_fear'] = df['NRC'].apply(get_NRC_fea)
df['NRC_pos_neg'] = df['NRC_positive'] - df['NRC_negative']
```
